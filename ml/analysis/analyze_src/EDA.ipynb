```python
# Step 4: Bivariate Analysis
# ------------------------------------------------
# Analyzing the relationship between two numerical features
# Gr Liv Area vs SalePrice
bivariate_analyzer = BivariateAnalyzer(NumericalVsNumericalAnalysis())
bivariate_analyzer.execute_analysis(df, 'Gr Liv Area', 'SalePrice')
```

The scatter plot above indicates a positive correlation between the Gr Liv Area and SalePrice i.e as Gr Liv Area increases, SalePrice also increases, suggesting that larger homes generally sell for higher prices. 
A majority of the data points are densely packed between 500 - 2500 sq.ft of living area and $50,000 - $300,000 in sale price, indicating that most homes fall within this size and price range. 
The scatter plot also shows some variations as Gr Liv Are increases. The spread of Sale Prices widens for larger homes, indicating that other factors besides size may influence home prices, such as location, quality, and additional features. 

```python
# Year built vs Sale Price 
bivariate_analyzer.execute_analysis(df, 'Year Built', 'SalePrice')
```

The above figure is a scatter plot of the sale price(y-axis) against the year built(x-axis). It can be observed from this plot that homes built before 1950 tend to have lower sale prices but with some variation. An increase in sale prices is evident after 1950, especially for homes built after 2000 and the highest-priced homes are clustered aroiund the newer homes(post-2000). Homes built after 2000 tending to have higher prices can indicate preference for newer constructions. Although some of these newer homes are selling for exceptionally high amounts which could be due to modern luxury homes, custom builds, or large lot sizes. There are some older houses with high sales prices as well, possibly due to location, size, or condition.
